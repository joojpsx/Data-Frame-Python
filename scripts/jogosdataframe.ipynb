# **DATAFRAME FEITO COM DICIONÁRIO**

```python
import pandas as pd #Importando a biblioteca "Pandas".

jogosdict = { #Criando o dicionário para derivar para Data Frame.
  "Nome": ["Valorant", "Red Dead Redemption 2", "Grand Chase", "Resident Evil 2 REMAKE", "Dota 2", "Dave The Diver"],
  "DataLan": ["02/06/2020", "26/10/2018", "XX/08/2003", "25/01/2019", "09/07/2013", "27/10/2022"],
  "Estúdio": ["Riot Games", "Rockstar Games", "KOG", "Capcom", "Valve Corporation", "MINTROCKET"],
  "Engine": ["Unreal Engine 4", "RAGE", "Gamebryo system", "RE Engine", "Source 2", "Unity"],
  "Gênero": ["FPS", "Ação e Aventura/Faroeste", "MMORPG", "Survival Horror", "MOBA", "RPG"]
}

jogos = pd.DataFrame.from_dict(jogosdict) #Criando o Data Frame a partir do dicionário.

jogos
```
